Let us consider a database containing the relative consumption of certain food items in European countries. The numbers represent the percentage of the population consuming that food type.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("https://openmv.net/file/food-consumption.csv").fillna(0)
data
```

```python
data.describe()
```

```python
A = data.to_numpy()[:,1:].T.astype(np.float64)
countries = data.to_numpy()[:,0]
foods = np.array(data.columns[1:])
```

```python
A
```

Each column of the matrix A refers to a country, and each row represents a specific food. The arrays countries
and foods contain the list of countries and foods, respectively.

1. How many features? How many samples?

```python
print('%d features' % A.shape[0])
print('%d samples'  % A.shape[1])
```

2. Perform PCA on the dataset by means of the SVD decomposition. Then, plot the trend of
- the singular values $\sigma_k$
- the cumulate fraction of singular values: $\frac{\sum_{i=1}^{k} \sigma_i}{\sum_{i=1}^{q} \sigma_i}$
- the fraction of the "explained variance": $\frac{\sum_{i=1}^{k} \sigma_i^2}{\sum_{i=1}^{q} \sigma_i^2}$

Comment on the results.

```python
np.mean(A, axis=1) # computes the mean of each row and put in an array with length equal to the number of rows
```

```python
A_norm = (A - np.mean(A, axis=1)[:,None]) / np.std(A, axis=1)[:,None]
```

```python
U, s, VT = np.linalg.svd(A_norm, full_matrices=False)

fig, axes = plt.subplots(nrows=1, ncols=3, figsize = (18,6))

axes[0].semilogy(s, 'o-')
axes[0].set_title('Singular Values')

axes[1].plot(np.cumsum(s)/np.sum(s), 'o-')
axes[1].set_title('Cumulate fraction of singular values')

axes[2].plot(np.cumsum(s**2)/np.sum(s**2), 'o-')
axes[2].set_title('Explained variance')
```

```python
Phi = np.matmul(U.transpose(), A_norm)
```

 Print the first two principal directions. Which are the more important foods for these two directions?
 Knowing that mediterranean cuisine envisages a large use of garlic and olive oil, compared to the European
 average, do you expect a positive or negative 1st principal component for Italy?

```python
print("First Principal Direction:", U[:, 0])
print("Second Principal Direction:", U[:, 1])
```

```python
# Print the food names with their weights for the first and second principal directions
print("Food contributions to the First Principal Direction:")
for i, food in enumerate(foods):
    print(f"{food}: {U[i, 0]:.3f}")

print("\nFood contributions to the Second Principal Direction:")
for i, food in enumerate(foods):
    print(f"{food}: {U[i, 1]:.3f}")
```

```python
first_direction_argmax = np.argmax(np.abs(U[:,0]))
second_direction_argmax = np.argmax(np.abs(U[:,1]))

print(f"1st direction food: {foods[first_direction_argmax]:}")
print(f"2st direction food: {foods[second_direction_argmax]:}")

```

```python
# Find the foods contributing the most to the first and second principal directions
first_principal_direction = [(food, U[i, 0]) for i, food in enumerate(foods)]
second_principal_direction = [(food, U[i, 1]) for i, food in enumerate(foods)]

# Sort by absolute value of contribution to identify the most influential
first_principal_direction_sorted = sorted(first_principal_direction, key=lambda x: abs(x[1]), reverse=True)
second_principal_direction_sorted = sorted(second_principal_direction, key=lambda x: abs(x[1]), reverse=True)

# Display the top contributing foods for each principal direction
print("Top Foods Contributing to the First Principal Direction:")
for food, contribution in first_principal_direction_sorted[:5]:
    print(f"{food}: {contribution:.3f}")

print("\nTop Foods Contributing to the Second Principal Direction:")
for food, contribution in second_principal_direction_sorted[:5]:
    print(f"{food}: {contribution:.3f}")
```

# **Cross-Checking Italy's Position in PC1**

To understand Italy's position in the first principal component (PC1), we compare:

1. **Principal Directions ($U[:, 0]$):**
   - $U[:, 0]$ describes how each food contributes to PC1.
   - Foods with **positive weights** increase PC1 if consumed more, while foods with **negative weights** decrease PC1.

2. **Italyâ€™s Food Consumption:**
   - Italy consumes:
     - **High amounts** of Garlic ($80$) and Olive Oil ($94$), which have **positive weights** in $U[:, 0]$.
     - **Low amounts** of Tinned Fruit ($9$), which has a **negative weight** in $U[:, 0]$.

3. **Expected Outcome:**
   - Since Italy aligns strongly with foods that positively contribute to PC1 (garlic, olive oil) and weakly with those that negatively contribute (tinned fruit), we expect **Italyâ€™s PC1 value ($\Phi[0, 1]$) to be positive**.

4. **Confirming with Data:**
   - Italyâ€™s PC1 value ($\Phi[0, 1]$) can be computed as the projection of Italyâ€™s normalized food data onto $U[:, 0]$.
   - This value represents Italyâ€™s alignment with PC1 and is expected to confirm the patterns observed in the original data.

```python
# 1. Print Crisp Bread's contribution to PC1 and PC2
crisp_bread_index = np.where(foods == "Crisp bread")[0][0]
print(f"Crisp Bread Contribution to PC1: {U[crisp_bread_index, 0]:.3f}")
print(f"Crisp Bread Contribution to PC2: {U[crisp_bread_index, 1]:.3f}")
print("")
# 2. Print Italy's PC1 and PC2 values
italy_pc1 = Phi[0, 1]
italy_pc2 = Phi[1, 1]
print(f"Italy's PC1: {italy_pc1:.3f}, PC2: {italy_pc2:.3f}")

# 3. Compare PC1 and PC2 values for all countries to Italy
print("\nCountries with food habits similar to Italy:")
for i, country in enumerate(countries):
    if i != 1:  # Skip Italy itself
        dist = np.sqrt((Phi[0, i] - italy_pc1)**2 + (Phi[1, i] - italy_pc2)**2)
        print(f"{country}: Distance={dist:.3f}")

# 4. Print England's PC1 and PC2 values
england_index = np.where(countries == "England")[0][0]
england_pc1 = Phi[0, england_index]
england_pc2 = Phi[1, england_index]
print(f"\nEngland's PC1: {england_pc1:.3f}, PC2: {england_pc2:.3f}")
```
