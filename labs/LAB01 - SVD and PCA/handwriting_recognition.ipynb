The MNIST dataset contains 28x28 pixel grayscale images of handwritten digits. When we load the data, these 2D images are flattened into 1D vectors of length 784 (28 * 28).

The `np.reshape()` function allows us to take these 1D vectors and reshape them back into their original 2D image format for visualization.

For example, let's say we have a 1D vector `image_data` that represents a single MNIST image. We can use `np.reshape()` to turn it back into a 2D 28x28 array like this:

```python
image_2d = image_data.reshape(28, 28)
```

This will take the 1D `image_data` vector and reshape it into a 2D 28x28 numpy array that we can then display as an image.

The key thing to remember is that the total number of elements has to be the same before and after the reshape. So a 1D vector of length 784 can be reshaped into a 2D array of shape (28, 28).

### Classification of two digits (0 and 9)

We now focus on the digits 0 and 9 only. Build a reduced training dataset containing only images corresponding to these digits.

````python
mask = np.logical_or(labels_full == 0, labels_full == 9)
A= A_full[:, mask]
labels = labels_full[mask]
````

Plot the first 30 training images. 

````python
fig, axs = plt.subplots(10,3, figsize=(5,8))
axs = axs.flatten()

for i in range(30):

    axs[i].imshow(np.reshape(A[:,i], (28,28)))
    axs[i].set_title("%d " %labels_full[i])
    axs[i].axis('off')
````

Compute and visualize the average of the training data

````python
avgImg = np.mean(A, axis =1)
reshapedAvgImage = np.reshape(avgImg, (28,28))
plt.imshow(reshapedAvgImage)
plt.axis('off')
````

Perform PCA on the data. Then, plot the trend of
- the singular values $\sigma_k$
- the cumulate fraction of singular values: $\frac{\sum_{i=1}^{k} \sigma_i}{\sum_{i=1}^{q} \sigma_i}$
- the fraction of the "explained variance": $\frac{\sum_{i=1}^{k} \sigma_i^2}{\sum_{i=1}^{q} \sigma_i^2}$

````python
U, s, VT = np.linalg.svd(A - avgImg[:, None], full_matrices = False)

fig, axs = plt.subplots(1,3, figsize=(18,6))
axs = axs.flatten()

axs[0].loglog(s)
axs[0].set_title("singular values")

axs[1].plot(np.cumsum(s)/np.sum(s))
axs[1].set_title("cumulative fraction of singular values")

axs[2].plot(np.cumsum(s**2)/np.sum(s**2))
axs[2].set_title("explained variance")

AS = U @ np.diag(s) @ VT
AS.shape
U[:, :3]
````

the plot of the singular values in semi-log scale shows a sudden drop to 0 after around 500 features. This is because the background pixels in the image data have no variance - the pixels are always the same black color across all the images.

The key insight here is that the intrinsic dimension of this image data is actually lower than the full 784 pixel dimensions, since the background pixels don't contain any useful information.

 the intrinsic dimension is lower than the total number of pixels (784) because some pixels are never used to represent the actual digit shapes.

So the solution is to focus on the principal components that capture the meaningful variation in the digit shapes, rather than trying to represent the full 784 pixel space

Visualize the first 30 principal axes (the columns of U)

````python
fig, axs = plt.subplots(10,3, figsize=(5,8))
axs = axs.flatten()

for i in range(1,31): 
    # Visualizzazione dell'immagine ricostruita
    axs[i-1].imshow(np.reshape(U[:,i], (28,28)))
    axs[i-1].set_title('k = %d' % i)  # Impostazione del titolo del subplot
    axs[i-1].axis('off')  # Rimozione degli assi
````

Principal component analysis tries to reconstruct the real object from the average plus a linear combination of the principal components

Compute the first two principal components corresponding to the first image

````python
first =np.inner(U[:,0],A[:, 0] - avgImg) 
second= np.inner(U[:,1],A[:, 0] - avgImg) 
print(first)
print(second)
````

The key idea is that the first principal component captures the most important variation in the data. And for representing the digit 0 versus the digit 9, the sign of the first principal component is very informative.

- If you want to represent a digit 0, you start from the average image and then you need to add (sum) a positive contribution from the first principal component. 
- In contrast, for representing a digit 9, you start from the average image and then you need to subtract (use a negative contribution) from the first principal component.

The instructor gives a great visual explanation for this:

"If you have a picture representing the zero, you expect the first principal component to be positive. Instead if you want to represent a 9, you start from here. OK? And do you want to sum or subtract this object? Subtract, so the first principal component would be negative."

So the sign of the first principal component indicates whether the image is more similar to a 0 (positive) or a 9 (negative). This provides a simple way to build a classifier just based on thresholding the first PC score.


Define a matrix corresponding all the principal components of all the images, and check the result of the previous point

````python
phi = U.T @ (A - avgImg[:, None])
print(phi[0,0])
print(phi[1,0])
````

Create a scatterplot for the first 2 principal components of all the training images, grouped by label. Consider only the first 500 pictures.

````python
for i in range(500):
    if labels[i] == 0:
        color = 'r'
    else:
        color ='b'
    first =np.inner(U[:,0],A[:, i] - avgImg) 
    second= np.inner(U[:,1],A[:, i] - avgImg)
    plt.scatter(first, second, color = color)
````

Define a classifier, by fixing a threshold on the first principal component. Then, test the classifier on the MNIST test dataset. Compute the number of true/false positive/negative for both the digits (confusion matrix) and the accuracy of the classifier.

We need a new dataset to use for prediction, and not the one used to train the pca

````python
data_test = np.genfromtxt('data/mnist_test.csv', delimiter=',')
labels_full_test = data_test[:,0]
A_full_test = data_test[:,1:].transpose()
labels_full_test.shape, A_full_test.shape
````

````python
mask = np.logical_or(labels_full_test == 0, labels_full_test == 9)
A_test = A_full_test[:, mask]
labels_test = labels_full_test[mask]
A_test.shape, labels_test.shape
````

````python
pc1_test = U[:,0] @ (A_test - avgImg[:, np.newaxis]) # I use the previous mean computed on the training dataset
labels_predicted = np.empty(labels_test.shape).astype(int)
labels_predicted[pc1_test < 0] = 9
labels_predicted[pc1_test >= 0] = 0

true_9 = np.sum(np.logical_and(labels_predicted == 9, labels_test == 9))
true_0 = np.sum(np.logical_and(labels_predicted == 0, labels_test == 0))
false_9 = np.sum(np.logical_and(labels_predicted == 9, labels_test == 0))
false_0 = np.sum(np.logical_and(labels_predicted == 0, labels_test == 9))

print('true 9: %d' % true_9)
print('true 0: %d' % true_0)
print('false 9: %d' % false_9)
print('false 0: %d' % false_0)
````

````python
print('accuracy: %1.2f percent' % ((true_9 + true_0)/(len(labels_test)) *100))
````
