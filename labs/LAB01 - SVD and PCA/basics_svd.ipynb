# SVD decomposition with `scipy`

Import the required modules

```
import scipy.linalg as la
import numpy as np
```

Generate a random 5x4 matrix

```
A = np.random.rand(5,4)
A
```

## `<numpy|scipy>.linalg.svd` 

Equivalent implementations:
- https://numpy.org/doc/stable/reference/generated/numpy.linalg.svd.html
- https://docs.scipy.org/doc/scipy/reference/generated/scipy.linalg.svd.html



### Full SVD (`full_matrices = True` - default)

**input:**

$A \in \mathbb{R}^{m\times n}$

**output:**

$U \in \mathbb{R}^{m\times m}, S \in \mathbb{R}^{m\times n}, V \in \mathbb{R}^{n\times n}$

or, more precisely

$\boldsymbol{\sigma} \in \mathbb{R}^{q} $

where $q = \min(m,n)$.


```
U, s, VT = np.linalg.svd(A)  #the third output is not v but v transpose
#U, s, VT = la.svd(A)
print('U shape: ', U.shape)
print('s shape: ', s.shape)
print('VT shape: ', VT.shape)
```

Build the matrix $S$

```
S = np.zeros(A.shape)
for i in range(len(s)):
    S[i, i] = s[i]
S
#here suppose you want to reconstruct the matrix sigma
# 1st way: manually- So initialize this matrix S with zeros of the same shape of A
# and them we manually put on the diagonal the singular values
```

```
S = la.diagsvd(s, A.shape[0], A.shape[1])
S
#2nd way: using this built-in function  that has exactly this purpose of reconstructing sigma from the singular values
#it takes exactly the singular values and the shape and we get sigma as a result
```

Reconstruct the matrix $A$

```
A_svd = np.matmul(U, np.matmul(S,VT)) # try if it works
# reconstruct A back from the SVD decomposition.
A_svd = U @ S @ VT # with @ whic is equal to matmul (same result)
print('err: %e' % (la.norm(A - A_svd) / la.norm(A))) #WE CHECK THE ERROR
```

### Thin SVD (`full_matrices = False`)

**input:**

$A \in \mathbb{R}^{m\times n}$

**output:**

$U \in \mathbb{R}^{m\times q}, S \in \mathbb{R}^{q\times q}, V \in \mathbb{R}^{n\times q}$

or, more precisely

$\boldsymbol{\sigma} \in \mathbb{R}^{q} $

where $q = \min(m,n)$.



```
del U, s, S, VT
U, s, VT = la.svd(A, full_matrices = False)
print('U shape: ', U.shape)
print('s shape: ', s.shape)
print('VT shape: ', VT.shape)
```

Build the matrix $S$

```
S = np.diag(s)
S
```

Reconstruct the matrix $A$

```
A_svd = np.matmul(U, np.matmul(S,VT))
print('err: %e' % (la.norm(A - A_svd) / la.norm(A)))
```
